# 4. Complete fundamental data quality and consistency checks.

```python
# Address data-types and potential mixed data-types in the dataframe
customers.dtypes
```

```python
# Checking for mixed data-types in each column
for col in customers.columns.tolist():
  weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# The 'first_name' column has mixed data types that need to be addressed
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# The 'user_id' column though, not containing mixed data types, is still improperly listed as integer
customers['user_id'] = customers['user_id'].astype('str')
```

```python
# Re-check data types in the data frame and for mixed data-types
customers.dtypes
```

```python
for col in customers.columns.tolist():
  weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

All columns are now appropriately updated to data types and there are no mixed data types present in any columns.
